# AAPL Stock Analysis

This project analyzes Apple (AAPL) stock using Python.

### What this project does:
- Collects real stock data using yfinance
- Visualizes stock price trends
- Calculates daily returns
- Measures risk (volatility)
- Evaluates cumulative performance

### Tools used:
Python, Pandas, Matplotlib, yfinance

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", start="2023-01-01", end="2024-01-01")
data.head()
```

```python
plt.plot(data["Close"])
plt.title("Apple Stock Price")
plt.show()
```

```python
data["Returns"] = data["Close"].pct_change()
data["Returns"].plot(title="Daily Returns")
plt.show()
```

```python
print("Average Daily Return:", data["Returns"].mean())
print("Risk (Std Dev):", data["Returns"].std())
```

```python
data["Cumulative Returns"] = (1 + data["Returns"]).cumprod()

plt.plot(data["Cumulative Returns"])
plt.title("Cumulative Returns of AAPL")
plt.show()
```
